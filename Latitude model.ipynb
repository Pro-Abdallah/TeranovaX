```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("asteroids_data.csv")
```

```python
data.head()
```

```python
data = data.drop(columns = ["Asteroid","Year","Energy_kilotons","Location","BlastRadius_km","Longitude","Velocity_km_s","AngularVelocity_deg_s"])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.show()
```

```python
y = data["Latitude"]
x = data.drop(columns = ["Latitude"])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression
mlat = LinearRegression()
mlat.fit(x_train,y_train)
```

```python
y_pred = mlat.predict(x_test)
```

```python
mae = mean_absolute_error(y_test,y_pred)
print("MAE: ",mae)
```

```python
df = pd.DataFrame({"y_test": y_test, "y_pred" : y_pred})
df
```
